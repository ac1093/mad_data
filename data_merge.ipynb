```python
url = 'https://en.wikipedia.org/wiki/List_of_wine-producing_regions'
url2 = 'https://en.wikipedia.org/wiki/List_of_countries_by_traffic-related_death_rate'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
```

```python
wine_producing_tables =pd.read_html(url)
traffic_tables = pd.read_html(url2)
wine_producing_df= pd.DataFrame(wine_producing_tables[0])
traffic_df =pd.DataFrame(traffic_tables[1])
wine_producing_df.head(10)
wine_producing_df = wine_producing_df.set_index("Country(with link to wine article)")
```

```python
traffic_table = pd.read_html(url2)
traffic_df = pd.DataFrame(traffic_table[1])
traffic_df = traffic_df.set_index("Country")
```

```python
# Create lists for country name and codes
name_list = []
country_codes = []
for country in pycountry.countries:
    name_list.append(country.name)
    country_codes.append(country.numeric)
# Return mismatching index from DF
for row, value in wine_producing_df.iterrows():
    if row not in name_list:
        print("ERROR: " + row)
```

```python
wine_producing_df.rename(index={
    "Russia": "Russian Federation",
    "Moldova": "Moldova, Republic of",
    "Macedonia": "North Macedonia",
    "Czech Republic": "Czechia",
    "Bolivia": "Bolivia, Plurinational State of",
    "Syria": "Syrian Arab Republic",
    "Reunion" : "Réunion"
    
}, inplace=True)
```

```python
#wine_producing_df = wine_producing_df.reset_index()
wine_producing_df
```

```python
# Create dict for countries
country_dict = dict(zip(name_list, country_codes))
country_dict
```

```python
# Create column to add to DF
new_col=[]
# Loop through indexes and get country_id
for row, value in wine_producing_df.iterrows():
    if row in country_dict:
        new_col.append(country_dict[row])
wine_producing_df['country_id'] = new_col
#wine_producing_df = wine_producing_df.set_index('country_id')
wine_producing_df
```

```python
wine_producing_df.reset_index(inplace = True)
wine_producing_df = wine_producing_df.rename(columns={'Country(with link to wine article)': 'country_name', 'Rank': 'rank', 'Production(tonnes)':'production(tonnes)'})
wine_producing_df.head(10) 
```
